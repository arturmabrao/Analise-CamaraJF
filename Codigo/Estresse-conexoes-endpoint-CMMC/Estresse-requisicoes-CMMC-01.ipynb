```python
## TESTE DE REQUISICAO - PAGINA DE LISTAGEM, SEM SLEEP, SEM TOLERÂNCIA DE FALHA

# VARIAVEIS DE CONTROLE
failed_req= 0
#failed = False
req_count = 0

# Variaveis de requisicao
num_proj = 1
ano_proj = 2015
tipo_proj = 'MOC'

while True:
    with requests.Session() as session:
        req_count+=1
        try:            
            # Requisicao para o portal de propostas da Camara
            query = {"numero": num_proj, "ano": ano_proj, "codtipo": tipo_proj}
            resposta = session.get('https://www.camarajf.mg.gov.br/sal/listapropos.php?', params=query)
        except:
            # Assinala que houve falha e quebra loop
            failed_req+=1
            break
            #failed = True
            #time.sleep(180)
                    
    session.close()
    #time.sleep(0.5)
    
print(f'Requisicoes feitas: {req_count}; \nRequisicoes com falha {failed_req}')
```
